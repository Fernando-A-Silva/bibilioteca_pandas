```python
import pandas as pd
import seaborn as sns
```

```python
dataset_varejo = pd.read_excel('varejo.xlsx')
dataset_cliente = pd.read_excel('cliente_varejo.xlsx')
```

```python
dataset_varejo.head()
```

```python
dataset_varejo.info()
```

```python
dataset_varejo.groupby('Nome_Departamento').idcompra.nunique()
```

```python
# prompt: quero substituir APP na base varejo por Aplicativo

dataset_varejo['Nome_Departamento'] = dataset_varejo['Nome_Departamento'].replace(' ', '_')
```

```python
# prompt: Quero substituir os espaços por _ nos valores da coluna Nome_Departamento da tabela dataset_varejo

dataset_varejo['Nome_Departamento'] = dataset_varejo['Nome_Departamento'].str.replace(' ', '_')
dataset_varejo.head()
```

```python
dataset_varejo[dataset_varejo['estado'].isnull()]
```

```python
#tratando valores nulos
varejo_snulos =  dataset_varejo.dropna()
varejo_snulos.info()
```

```python
#substituindo nulos
dataset_varejo['estado'].fillna('MS', inplace=True)
dataset_varejo.info()
```

```python
#substituindo valores nulos de preço pela média dos preços da base
media_preco = dataset_varejo['Preço'].mean()

dataset_varejo['Preço'].fillna(media_preco, inplace=True)
dataset_varejo.info()
```

```python
dataset_varejo[['Preço', 'Preço_com_frete']].describe()
```

```python
#Identificando preços incorretos

preco_incorreto = dataset_varejo.query('Preço > Preço_com_frete')
preco_correto = dataset_varejo.query('Preço < Preço_com_frete')
```

```python
preco_correto.head()
```

```python
preco_correto.query("estado == 'SP' and Nome_Departamento == 'Esporte_e_Lazer'").describe()
```

```python
#departamentos mais vendidos
preco_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#média de preço de departamentos mais vendidos
round (preco_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
#qtd de vendas por data
preco_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#extraindo mês de uma data e criando nova coluna
preco_correto['Mês'] = preco_correto['Data'].dt.month
```

```python
#qtd de vendas por mês
preco_correto.groupby('Mês').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#tratando base de clientes

dataset_cliente = dataset_cliente.astype({'renda': 'float'})
dataset_cliente.info()
```

```python
# prompt: Quero fazer uma união das bases varejo e cliente utilizando como chave a coluna cliente_Log

dataset_varejo_cliente = pd.merge(dataset_varejo, dataset_cliente, on='cliente_Log', how='left')
dataset_varejo_cliente.head()
```

```python
#media de renda por canal de venda

agg_idcanal_renda = round( dataset_varejo_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
#media de idade por bandeira

agg_idade_bandeira = dataset_varejo_cliente.groupby('bandeira').idade.mean().sort_values(ascending=False).reset_index()
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize= (7,4))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'])
plt.title('Média de idade por bandeira', loc = 'center', fontsize =18, fontweight = 'bold', color='black')
plt.xlabel('Bandeira')
plt.ylabel('Média de idade');
```

```python
fig = plt.figure(figsize= (7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'])
plt.title('Média de renda por canal de venda', loc = 'center', fontsize =18, fontweight = 'bold', color='black')
plt;
```

```python
#vendas ao longo do tempo

vendas_por_data = preco_correto.groupby('Data').idcompra.nunique().reset_index()

fig = plt.figure(figsize= (20,12))

plt.xlabel('Data')
plt.ylabel('Vendas')
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']

plt.plot(x,y, linewidth = '2')
plt.title('Vendas ao longo do tempo', loc = 'center', fontsize =18, fontweight = 'bold', color='black');
```

```python
agg_mediapreco_departamento = round (preco_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)

fig = plt.figure(figsize= (30,20))

plt.bar(agg_mediapreco_departamento['Nome_Departamento'], agg_mediapreco_departamento['Preço_com_frete'])
plt.title('Média de preço por departamento', loc = 'center', fontsize =18, fontweight = 'bold', color='black')
plt.xlabel('Departamento')
plt.ylabel('Média de preço');
```
